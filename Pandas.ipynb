**Pandas Series**

```python
import pandas as pd
```

```python
series1= pd.Series([1,2,3,4,"Shan"])
```

```python
print(series1)
```

```python
series1
```

```python
#Find out the datatype of Series1
type(series1)
```

```python
#Create a Python List of the Weights of 5 Students of the class and convert it into a Pandas Series,Check Whether it was converted it successfully or not
weight=[60,70,80,90,85]
weight_series= pd.Series(weight)
print(weight_series)
```

```python
type(weight_series)
```

```python
#Find out the size of the Pandas Series
weight_series.size
```

```python
#To Find out the Shape
print("The Shape of series is",weight_series.shape)
```

```python
#Find out Mean,Min and Max
print("The mean value is",weight_series.mean())
print("The maximum value is",weight_series.max())
print(weight_series.min())
```

```python
#Display Rows using head and tail Functions:
#head
print("Top Five rows",weight_series.head())
print("Top three rows",weight_series.head(3))
```

```python
#tail
print("Bottom Five Rows",weight_series.tail())
print("Bottom three rows",weight_series.tail(3))
```
